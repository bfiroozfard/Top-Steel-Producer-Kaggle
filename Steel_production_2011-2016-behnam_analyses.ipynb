```python
# remove ranking from data 
a=db.groupby(db.Headquarters).agg(sum)
b=pd.DataFrame(a)
```

```python
c=b.iloc[:,[0,1,2,3,4,5]].copy()
c
```

```python
 # add average column
 c['ave_production']=c.mean(axis=1)
```

```python
 # add sum column
c['sum production']=c.iloc[:,[1,2,3,4,5,6]].sum(axis=1)
```

```python
c.sort_values(['ave_production'],ascending=False,inplace=True)
```

```python
c.head()
```

```python
c.head()
```

```python
plt.figure(figsize=(15, 3))
plt.plot(c.index, c.ave_production,'bx--', label='line chart ',)  # Plot some data on the (implicit) axes.
plt.xlabel('Head Quarters')
plt.ylabel('Average Steel Production /million Tonnage ')
plt.title("Top 100 Steel producer")
plt.xticks(rotation=30, fontsize='small')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 6))

plt.subplot(211)
plt.bar(c.index, c.ave_production)
plt.xticks(rotation=30, fontsize='small')

plt.subplot(212)
plt.scatter(c.index, c.ave_production)
plt.xticks(rotation=30, fontsize='small')

plt.show()
```
